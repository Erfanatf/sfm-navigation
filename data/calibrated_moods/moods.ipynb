**How to read the numbers:**  
- **Kinematic** cluster is determined mainly by `speed_std`, `accel_*`, `jerk_*` (not all shown here; the table includes speed metrics). Notice how “Energetic” (label 1) regimes have higher speed std, lower stop ratio.  
- **Path** cluster: “Straight” regimes have tortuosity ~1.0–1.3 and low curvature values. “Winding” ones have tortuosity >10 with huge max curvature.  
- **Safety** cluster: “Isolated” has very low `intrusion_time_frac` and zero `num_intrusions`. “Crowded” has higher `intrusion_time_frac` (~0.4–0.7) but still zero `num_intrusions`. “Intrusive” has `num_intrusions` > 0 (discrete intrusions).  
- **Social** cluster: “Low‑attention” has `fov_occupancy` ~5–7, `mutual_attention_count` ~1.5–2. “High‑attention” has ~18–20 fov occupancy and ~7 mutual attention. “Group‑affiliated” has high `group_affiliation` (0.18–0.26).

These raw feature means are exactly the “degree of belonging” – a window is assigned to the cluster whose centroid (after PCA) is closest, but these average values explain the typical behaviour of that regime.
